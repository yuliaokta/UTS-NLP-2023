```python
# Tokenisasi kata-kata
words = word_tokenize(text)
```

```python
# Menghapus stopwords (kata-kata umum yang tidak relevan)
stop_words = set(stopwords.words('indonesian'))
filtered_words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]
```

```python
# Menghitung frekuensi kata-kata
fdist = FreqDist(filtered_words)
```

```python
# Mengambil 20 kata yang paling sering muncul
top_words = fdist.most_common(20)
```

```python
# Membuat DataFrame untuk plot
df = pd.DataFrame(top_words, columns=['Kata', 'Frekuensi'])
```

```python
# Membuat plot bar
plt.figure(figsize=(10, 6))
plt.bar(df['Kata'], df['Frekuensi'])
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.title('Top 20 Kata yang Paling Sering Muncul')
plt.xticks(rotation=45)
plt.show()

# Analisis singkat
print("Kata-kata yang paling sering muncul dalam teks ini adalah:")
for index, row in df.iterrows():
    print(f"{row['Kata']}: {row['Frekuensi']} kali")

print("\nDengan analisis ini, dapat mengidentifikasi kata-kata kunci yang paling relevan dalam teks pidato  Pidato HUT RI, Ganjar: Ideologi Negara Tak Bisa Diganti Sampai Kapanpun, yang dapat membantu memahami topik atau tema utama dalam teks pidato tersebut.")
```
